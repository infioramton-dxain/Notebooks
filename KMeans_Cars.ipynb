```python
import pandas as pd
import numpy as np

import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.figure_factory as ff
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('Data_PCA.csv')
```

```python
dummies = pd.get_dummies(data.CARID, prefix='CARID')
data = pd.concat([data, dummies], axis=1)
data.drop(['CARID'], axis=1, inplace=True)
```

```python
mms = MinMaxScaler()
mms.fit(data)
data_transformed = mms.transform(data)
```

```python
Sum_of_squared_distances = []
K = range(1,20)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
km.inertia_
```

```python
SSDdf = pd.DataFrame(Sum_of_squared_distances)
SSDdf.rename(columns = {0:'SSD'}, inplace = True)
SSDdf['Percentage'] = 100*SSDdf.SSD.cumsum()/SSDdf.SSD.sum()
```

```python
fig = px.scatter(SSDdf,
                x = K,
                y = SSDdf.SSD,
                hover_name=SSDdf.Percentage)
fig.show()
```

```python
#View the Scree plot to identify the elbow in the curve
fig = make_subplots(rows=2, cols=1)
fig.add_trace(
    go.Bar(
        x = SSDdf.index, 
        y = SSDdf.Percentage),
        row=1, col=1
)

fig.add_trace(
    go.Scatter(
        x = SSDdf.index, 
        y = SSDdf.SSD),
        row=2, col=1
)

fig.update_layout(height=500, width=1000, title_text="Scree Plots")
fig.show()
```
