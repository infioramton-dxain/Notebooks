```python
#Import Libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ForAsst4Country.csv')
```

# THE SORTING SECTION

```python
#Sort on One column, Ascending
df.sort_values(['Country Name'], ascending=True)

#Sort on One column, Descending
df.sort_values(['Country Name'], ascending=False)

#Sort on Multiple column, varying ASC/DESC
df.sort_values(['Country Name', 'International Category Name', 'Fiscal Year'], ascending=[True, True, False])

#Using quicksort or mergesort (more stable algorithm)
df.sort_values(['Country Name', 'Current Dollar Amount'], ascending=True, kind='mergesort')

#Sorting column labels
df.columns.sort_values()

#Sort on One column, but bring any NANs to the top
df.sort_values(['Country Name'], ascending=True, na_position="first")

#Sort on One column, and update the dataframe at the same time to keep the data sorted
df.sort_values(['Country Name'], ascending=True, inplace=True)

#Sort on the Index if it was Sorted on a column inplace before.
df.sort_index(inplace=True)

#Sort on a column and then select only the relevant columns.  
#This way you can group/pivot on the data you want instead of the full dataset.
df.sort_values(['Country Name', 'Fiscal Year'])[['Country Name', 'Fiscal Year', 'Current Dollar Amount']]

#Sort by the column name alphabetically
df.sort_index(axis = 1)
```

# THE GROUP SECTION

```python
#Group by a single column/category.  Count records by grouped category.
df.groupby(["Country Name"]).count()

#Group by a category, and sum by another
df.groupby(["Country Name"])["Current Dollar Amount"].sum()

#Group by two categories, and sum by another
df.groupby(["Country Name", "Fiscal Year"])["Current Dollar Amount"].sum()

#Group by two categories, and sum by another, then Rename the Columns
df.groupby(["Country Name", "Fiscal Year"])["Current Dollar Amount"].sum().rename_axis(["A", "B"])

#Group by two categories, using aggregate function
df.groupby(["Country Name", "Fiscal Year"]).agg({'Current Dollar Amount':sum})
df.groupby(["Country Name", "Fiscal Year"]).agg({'Current Dollar Amount':min})
df.groupby(["Country Name", "Fiscal Year"]).agg({'Current Dollar Amount':max})

#Group by two categories, with multiple aggregate function on another category
df.groupby(["Country Name", "Fiscal Year"])['Current Dollar Amount'].agg(['sum','min','max'])

#Group by two categories, with an aggregate function on another category, and SORTED
df.groupby(["Country Name", "Fiscal Year"])['Current Dollar Amount'].sum().sort_values(ascending=False)
df.groupby(["Country Name", "Fiscal Year"])['Current Dollar Amount'].max().sort_values(ascending=False)


#Group by two categories, Sum, Sort, and get the top/bottom value
df.groupby(["Country Name", "Fiscal Year"])['Current Dollar Amount'].sum().sort_values(ascending=False).head(1)
df.groupby(["Country Name", "Fiscal Year"])['Current Dollar Amount'].sum().sort_values(ascending=False).tail(1)
```

# THE PIVOT SECTION

```python
#Pivot (group by) on Country/Category, Dimensions on Fiscal Year, Aggregate on Dollar Amount
pd.pivot_table(df,
               index=["International Category Name"],
               values=["Current Dollar Amount"], 
               columns=["Country Name"],
               aggfunc=[np.sum],
               fill_value=0)

#Pivot and select amount fields for different processes. Assume Fiscal Year is a summable field like CDAmount.
pd.pivot_table(df, 
                index=["International Category Name"],
                values=["Current Dollar Amount","Fiscal Year"],
                columns=["Country Name"],
                aggfunc= {"Current Dollar Amount":np.sum,
                        "Fiscal Year": [np.sum, np.mean]})
```

```python
#Pivot (group by) on Country/Category, Dimensions on Fiscal Year, Aggregate on Dollar Amount
dfpivot = pd.pivot_table(df, 
               index=['Country Name','International Category Name'], 
               columns=['Fiscal Year'], 
               values=['Current Dollar Amount'],
               aggfunc={'Current Dollar Amount':[np.sum, np.mean, np.min, np.max]},
                fill_value=0)

#Format the Pivot table.
dfpivot.style. \
    background_gradient(cmap='Blues', axis=None). \
    set_table_styles(
    [{'selector': 'th,td,tr', 'props': [('border-style', 'solid'), ('border-width', '1.0px')]}])
```

# THE FILTER SECTION

```python
#Filter the dataset for records that have Country Afghanistan.  Returns a set of index and true/false
df["Country Name"]=="Afghanistan"

#Filter the dataset but return a dataset and show just the top two records.
df[df["Country Name"]=="Afghanistan"].head(2)

#Filter the dataset for records with funding over $5M and show top two.
df[df["Current Dollar Amount"] > 5000000].head(2)

#Combine the filters and show LAST two records
df[(df["Current Dollar Amount"] > 5000000) & (df["Country Name"] == "Afghanistan")].tail(2)

#Filter across a set of values.
df[df['Country Name'].isin(['Afghanistan', 'Albania'])]

#Combine Filter for Country and then Pivot and sum
dfAFG = pd.pivot_table(df[df["Country Name"]=="Afghanistan"], 
               index=['Country Name','International Category Name'], 
               columns=['Fiscal Year'], 
               values=['Current Dollar Amount'],
               aggfunc={'Current Dollar Amount':[np.sum]},
                fill_value=0)

#Or filter before you use the pivot
df[df["Country Name"]=="Afghanistan"].\
            pivot_table(index=['Country Name','International Category Name'], 
            columns=['Fiscal Year'], 
            values=['Current Dollar Amount'],
            aggfunc={'Current Dollar Amount':[np.sum]},
            fill_value=0)
```

```python
#Combine Filter for Country and then Pivot and sum.  Changed from np.sum because of error.
dfAFG = pd.pivot_table(df, 
               index=['Country Name','International Category Name'], 
               columns=['Fiscal Year'], 
               values=['Current Dollar Amount'],
               aggfunc='sum',
                fill_value=0)

#Show which Categories have Negative values or above $60M
dfAFG[(dfAFG["Current Dollar Amount"]<0) | (dfAFG["Current Dollar Amount"]>600000000)]
```
